```plaintext
import seaborn as sns
```

```plaintext
iris = sns.load_dataset('iris')
```

```plaintext
iris
```

```plaintext
sns.scatterplot(x="sepal_length",y="sepal_width",data = iris)
```

```plaintext
sns.lineplot(x = "petal_length",y = "petal_width",data = iris)
```

```plaintext
sns.distplot(iris["petal_length"])
```

```plaintext
sns.countplot(x = "petal_width",data = iris)
```

```plaintext
sns.boxplot(x="sepal_length",y="sepal_width",data = iris)
```

```plaintext
iris.describe()
```

```plaintext
sns.catplot(x="sepal_length",y= "sepal_width",hue = "species",kind = "bar",data = iris)
```

```plaintext
sns.pairplot(iris)
```

```plaintext
iris.corr()
```

```plaintext
sns.heatmap(iris.corr(),annot= True)
```
