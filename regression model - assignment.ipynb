```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("50__startups.csv")
```

```python
dataset["Administration"].fillna(dataset["Administration"].mean(),inplace = True)
dataset["Marketing Spend"].fillna(dataset["Marketing Spend"].mean(),inplace = True)
```

```python
dataset.isnull().any()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset["State"]=le.fit_transform(dataset["State"])
```

```python
x=dataset.iloc[:,0:4].values
y=dataset.iloc[:,4:].values
```

```python
type(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
Regression_model = Sequential()
```

```python
Regression_model.add(Dense(units=6,kernel_initializer="normal",activation="relu"))
```

```python
Regression_model.add(Dense(units=180,kernel_initializer="normal",activation="relu"))
```

```python
Regression_model.add(Dense(units=180,kernel_initializer="normal",activation="relu"))
```

```python
Regression_model.add(Dense(units=10,kernel_initializer="normal",activation="linear"))
```

```python
Regression_model.compile(optimizer="adam",loss="mean_squared_error",metrics=['mean_squared_error'])
```

```python
history=Regression_model.fit(x_train,y_train,epochs=1000,batch_size=128,validation_data=(x_test,y_test))
```

```python
y_pred=Regression_model.predict(x_test)
```
